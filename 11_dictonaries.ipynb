# Dictonary

```python
# Dictionary is mutable/ change-able  data structure that store data in key value pairs.
```

```python
students = {
    'name' : 'Adnan',
    'roll no' : 22,
    'department' : 'cs'
}

#Keys
students.keys()

```

```python
#values
students.values()
```

```python
#items
students.items()
```

```python
# copy the dictonary
new_students = students.copy()
new_students
```

```python
# get
students.get('name')
```

```python
#get
students.get('class')
# it show null value

students.get('class','not available')
```

```python
# pop
students.pop('name')
```

```python
students
```

```python
# push
students.update({'name':'adnan'})
students
```

```python
#pop item remove  the last item
students.popitem()
```
