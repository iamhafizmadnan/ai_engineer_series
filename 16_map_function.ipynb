# Map Function


```python
# map , reduce and filter
a = 5

def square(x):
  return x **2

print (square(a))
```

```python
alist = [2,4,5,6,7,9]
def square(x):
  return x **2

print (list(map(square,alist)))
```

```python
# map function
alist = [2,4,6,7,8,9]

print(map(lambda p:p*p , alist))

print (list(map(lambda p:p*p , alist)))
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))
print(odd_numbers)
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
odd_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(odd_numbers)
```

```python
#reduce function
from functools import reduce
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)
```
