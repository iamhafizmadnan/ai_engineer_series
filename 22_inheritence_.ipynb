# Inheritence

```python
class A:

  def method1(self):
    print('This is a Method1')

  def method2(self):
    print('This is a Method2')


class B(A):
  #  child class/derived class

  def method3(self):
    print('This is a Method3')


  def method4(self):
    print('This is a Method4')


class C(B): # Changed from C(A,B) to C(B)
  def method5(self):
    print('This is a Method5')


  def method6(self):
    print('This is a Method6')



b = B()
b.method3()
b.method4()
b.method1()
b.method2()

c = C()
c.method1()
c.method2()
c.method3()
c.method4()
c.method5()
```
