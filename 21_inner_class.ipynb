# inner class



```python
class employee:
  def __init__(self, name, role, salary):
    self.name = name
    self.role = role
    self.salary = salary

  def employee_details(self):
    print(f"{self.name} is a {self.role} with a salary of {self.salary}")

  # inner class
  class department:
    def __init__(self, street, city):
      self.street = street
      self.city = city

    def department_details(self):
      print(f"{self.street} is a {self.city}")

# These lines should be outside the class definition
employee1 = employee("John", "Manager", 50000)
employee1.employee_details()



department1 = employee.department("123 Main St", "New York")
department1.department_details()
department2 = employee.department("456 Elm St", "Los Angeles")
department2.department_details()
```
