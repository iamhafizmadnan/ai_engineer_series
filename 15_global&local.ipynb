# Global and local variable

# Local variable
A variable created inside a function belongs to the local scope/variable of that function, and can only be used inside that function.



```python
def my_func():
  name = 'adnan'
  print(name)

my_func()
```

```python
def my_func():
  name = 'adnan'
  def inner_func():
    print(name)
  inner_func()

my_func()
```

# Global variable
A variable created in the main body of the Python code is a global variable and belongs to the global scope.

Global variables are available from within any scope, global and local.

```python
x = 300

def myfunc():
  print(x)

myfunc()  #local

print(x)  # global
```

```python
x = 300

def myfunc():
  x = 200
  print(x)

myfunc()

print(x)
```

# Lambda Function
A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.

```python
x = lambda a : a + 10
print(x(5))
```

```python
x = lambda a,b : a * b
print(x(4,6))
```
