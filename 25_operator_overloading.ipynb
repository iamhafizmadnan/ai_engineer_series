# Operator Overloading

operator overloading in python allows you to redefine how built_in_operator behave
### for user defined object (classes). This can be achieved by special functions like add subtration etc.

```python
#addition
x = 5
y = 8

print(int.__add__(x,y))
```

```python
# concatenation
x = '5'
y = '8'
print(x+y)
```

```python
class invoice:
  def __init__(self, invoiceamount):
    self.amount = invoiceamount

  def __add__(self, other):
    total =  self.amount + other.amount
    print(total)

invoice1 = invoice(100)
invoice2 = invoice(200)
invoice1 + invoice2
```

```python
class invoice:
  def __init__(self, invoiceamount):
    self.amount = invoiceamount

  def __sub__(self, other):
    total =  self.amount - other.amount
    print(total)

invoice1 = invoice(300)
invoice2 = invoice(200)
invoice1 - invoice2
```
