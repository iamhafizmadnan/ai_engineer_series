# While Condition

```python
# while condition:
   # Do Something
   # Do Something
x = 5
while x < 9:
  print ("Hello")
  x = x + 1
```

```python
# for loop --  iterate over a sequence like range, string ,list etc.
for i in range (5):
  print(f'Hello: {i}' )
```

# For Loop

```python
for i in range (5, 20):
  print(f'Hello: {i}' )
```

```python
for i in range (5,20,5):
  print(f'Hello: {i}' )
```

```python
for i in range (5,20):
  if i == 15:
    continue
  else:
      print(f'Hello: {i}' )

```

```python
for i in range (2,10):
  if i == 7:
    break
  else:
      print(f'Hello: {i}' )
```

```python
fruits = ['apple','orange','grapes']
for c in fruits:
  print(c)
```

```python
# display tabel of 7 number
for i in range (1,11):
  print(f'7 x {i} = {7*i}')
```

```python
#write a program to get factorial of 5
fact = 1
for i in range (1,6):
  fact = fact * i
print(fact)

```
