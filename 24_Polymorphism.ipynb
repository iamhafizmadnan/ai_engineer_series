## Polymorphism (Many Types)

### Duch Type
Object must have the minimum necessary attributes/methods to consider them of certain types.
If looks like duck and quacks like duck ,it must be a duck
1. Operator Overloading
2. Method Overloading
3. Method Overriding

```python
class vehicle():
  run = True


class car(vehicle):
  def color(self):
    print("This car has  black color")

class bike(vehicle):
  def color(self):
    print("This bike has  red color")


class dog:
  def color(self):
    print("This dog has  black color")

class cat(dog):
  def color(self):
    print("This cat has  white color")
vehicles = [car(), bike(), cat(), dog()]
for vehicle in vehicles:
  vehicle.color()



```
