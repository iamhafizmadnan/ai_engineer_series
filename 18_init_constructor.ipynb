## parameterized construcotr


```python
class student:
  def __init__(self,name,id):

    # attribute
    self.name=name
    self.id=id

    # methods
  def display(self):
      print(self.name,self.id)



adnan = student("adnan",123)
print(adnan.name)
print(adnan.id)
```
