```python
# class variable and instance class


class employee:
  '''This is used for class variable and instance class'''

  #class / static variable
  office = 'pakistan'
  time = '9 to 5'


  def __init__(self, name , salary, role):
    # attribute / instance classes
    self.name = name
    self.salary = salary
    self.role = role

    #method
    def printdetails(self):
      return f"Name is {self.name}. Salary is {self.salary} and role is {self.role}"

emp1 = employee("Ali", 50000, "Developer")
emp2 = employee('zain', 40000, "Manager")

print(emp1)
print(emp2)

# __doc__
print(employee.__doc__)

#__dict__
print(emp1.__dict__)
print(emp2.__dict__)


#__name__
print(employee.__name__)

#__module__
print(employee.__module__)

#__bases__
print(employee.__bases__)


```
