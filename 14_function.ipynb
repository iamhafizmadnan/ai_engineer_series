# Function


```python
# Arguments
def my_fun(fname): # this is the argument inside the paranthesis.
  print(fname + ' Sami')

my_fun('Adnan')
my_fun('Ali')
my_fun('zain')
```

```python
# Two arrguments
def my_fun(fname, Lname):
  print(fname +' ' +Lname)

my_fun('Adnan', ' Sami')

```

```python
# Two arrguments show error when you give only one argument.

def my_fun(fname, Lname):
  print(fname +' ' +Lname)

my_fun(' Sami')
```

```python
# default parameter
def my_fun(country = 'Pakistan'):

  print("i am from "  + country)

my_fun("uk")
my_fun("USA")
my_fun()
my_fun("UAE")
```

```python
# keyword argument
def my_fun(child3, child2, child1):
  print("The youngest child is " + child3)

my_fun (child1 = "Emil", child2 = "Tobias", child3 = "Linus")
```

```python
# position argument
def my_fun(child3, child2, child1):
  print("The youngest child is " + child3 , child1, child2)

my_fun("Emil", "Tobias", "Linus")
```

```python
# mix positional and keyword arguement
def my_fun(name, age, city):
  print("My name is", name , "And my age is", age, 'and my city is', city)

my_fun("Adnan" , age = 33, city = "karachi")
```

```python
# passing different value
def my_fun(city):
    for city in city:
        print(city)

city = ("karachi", "lahore", "islamabad")

my_fun(city)
```

```python
# return value
def my_fun(x):
  return 5 * x

print(my_fun(3))
```

```python
def my_func(x,y):
  return x * y

result =my_func(8 ,8)
print(result)
```

```python
# returning data types
def my_fun():
  return ['apple', 'orange','bannan']
fruits = my_fun()

print(fruits[0])
print(fruits[1])
print(fruits[2])
```

```python
# Keyword-Only Arguments
def my_fun(a, b, *, c):
    print(a, b, c)

my_fun(1, 2, c=3)

```

```python
# arbitrary argument
def my_fun(*kids):
  print("The youngest child is " + kids[2])

my_fun ('adnan', 'sami', 'ali')
```
