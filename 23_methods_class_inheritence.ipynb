# Methods class in inheritence

```python
class A:

  def __init__(self):
    print("A init is here")

def method1():
    print("A method1")

def method2():
    print("A method2")

def method3():
    print("A method3")

# child class
class B(A):

  def method1():
    print("B method1")

  def method4():
    print("B method4")


b = B()
```

```python
class A():

  def __init__(self):
    print("A init is here")


  def method1():
    print("A method1")

  def method2():
    print("A method2")


# Moved class B outside of A and made it inherit from A
class B(A): # B now explicitly inherits from A

    def __init__(self):
      print("B init is here")
      super().__init__() # This will now call A.__init__


    def method1():
     print("B method1")

    def method2():
     print("B method2")


class C(B): # C inherits from A and B(A)

  def __init__(self):
    print("C init is here")
    # super().__init__() # This will call B.__init__ based on MRO: C -> B -> A -> object


  def method1():
    print("C method1")

  def method2():
    print("C method2")


c = C()
```
