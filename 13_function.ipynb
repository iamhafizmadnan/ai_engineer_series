# Functions
A function is a block of code which only runs when it is called.

A function can return data as a result.

A function helps avoiding code repetition.

```python
# Two types of functions
 # Built_in_funciton
 # User_defined_function
```

```python
#Built_in_function
print('Hello Adnan')
```

```python
#user definded function
a = 23
b = 32
z = a + b
print(z)
```

```python
def function():
  a = 22
  b = 99
  z = a + b
  print(z)
function()
```

```python
def addition(a,b):
  z = a+b
  print(z)
a = 44
b = 89

addition(a,b)
```

```python
book = 100
tax = 0.09
total = book * tax
print(total)
```

```python
def taxcal(book, tax):
  total = book * tax
  print(total)
tax = 0.07
book = 50

taxcal(book, tax)
```
