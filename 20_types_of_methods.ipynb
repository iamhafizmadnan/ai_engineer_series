```python
class employee:
  '''This is used for class variable and instance class'''

  #class / static variable
  office = 'pakistan'
  time = '9 to 5'


  def __init__(self, name , salary, role):
    # attribute / instance variable
    self.name = name
    self.salary = salary
    self.role = role

  # 1 instance method
  def displaystudent(self):
    return f"Name is {self.name}. Salary is {self.salary} and role is {self.role}"

# 2 class method
  @classmethod
  def changeoffice(cls):
    print(f"Office is {cls.office}")

# 3 static method
  @staticmethod
  def statmethod():
    print("This is static method")

#static method

  #method
  def printdetails(self):
    return f"Name is {self.name}. Salary is {self.salary} and role is {self.role}"

emp1 = employee("Ali", 50000, "Developer")
emp2 = employee('zain', 40000, "Manager")



print(emp1.displaystudent())

print(emp2.displaystudent())

print(emp1.office)

print(emp2.office)

print(emp1.statmethod())

print(emp2.statmethod())
```
